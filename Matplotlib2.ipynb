```python
# Matplotlib Part - 2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generate a graph using a Dataset
data = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/Grade_Set_1.csv')
data
```

```python
plt.scatter(data.Hours_Studied, data.Test_Grade, color='red')
plt.plot(data.Hours_Studied, data.Test_Grade, color='green', label='Data From Dataset')
plt.xlabel('Hours_Studied', fontsize=12)
plt.ylabel('Marks', fontsize=12)
plt.title('Hours_Studied vs Test Grade', fontsize=10, fontweight='bold')
plt.legend()
plt.show()
```

```python
#xticks() and yticks()
from numpy import random
x = random.randint(100,size=(10))
y = random.randint(60,size=(10))
plt.plot(x,y,color='red')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Xticks and Yticks')
plt.legend()
plt.show()
```

```python
#xticks() and yticks()
from numpy import random
x = random.randint(100,size=(10))
y = random.randint(60,size=(10))
plt.plot(x,y,color='red')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Xticks and Yticks')
plt.xticks(np.arange(0,105,5))
plt.yticks(np.arange(0,60,5))
plt.show()
```

```python
# Histogram
# a histogram is used to show the distrubution of data
# for example : if you have the ages of 100 people , a histogram help you see how many people are in their in 20's
# 30s, 40s
# ages = [18,21,22,25,26,30,34,35,36,40,42,45,46,50]
# bins = 5
# divide the range of ages(18-50) into 5 equal sized groups
# 18-24 25-31 32-38 39-45 46-52

ages = [18, 21, 22, 25, 26, 30, 34, 35, 36, 40, 42, 45, 46, 50]
plt.hist(ages, bins=5, color='skyblue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of People")
plt.show()
```

```python
# Heatmap
# a heatmap is a colorful table
# instead of showing numbers as text , it will show you colors to make patterns
# easier to spot

# a tables where each cell's color tells you "how big or small" a value is
# darker color for higher value
# lighter color for smaller value

# correlation
# correlation used to find out the relationship

# correlation is measured on a scale of +1 and -1

# imagine you 2 things  a.) Hours Studied  b.) Marks scored
# if a student studies more hours usually more marks  : pos relationship
# if student watches more TV and scores less          : neg relationship


data = {
    'Hours_Studied':[2,3,4,5,6,7,8],
    'Marks_Scored':[40,50,60,70,80,90,100],
    'TV_Watched':[10,9,8,6,7,5,3]
}

df = pd.DataFrame(data)
df.corr()
```

```python
# Heatmap
# a heatmap is a colorful table
# instead of showing numbers as text , it will show you colors to make patterns
# easier to spot

# a tables where each cell's color tells you "how big or small" a value is
# darker color for higher value
# lighter color for smaller value

# correlation
# correlation used to find out the relationship

# correlation is measured on a scale of +1 and -1

# imagine you 2 things  a.) Hours Studied  b.) Marks scored
# if a student studies more hours usually more marks  : pos relationship
# if student watches more TV and scores less          : neg relationship


import seaborn as sns


data = {
    'Hours_Studied': [2, 3, 4, 5, 6, 7, 8],
    'Marks_Scored': [40, 50, 60, 70, 80, 90, 100],
    'TV_Watched': [10, 9, 8, 6, 7, 5, 3]
}

df = pd.DataFrame(data)
corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation')
plt.show()
```
